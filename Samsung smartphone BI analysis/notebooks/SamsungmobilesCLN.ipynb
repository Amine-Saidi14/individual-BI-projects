Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", None)
```

Load Dataset

```python
df = pd.read_csv("samsungMobilesData.csv")

df.head()
```

Basic Dataset Inspection

```python
df.shape
df.info()
df.describe()
```

Check Missing Values

```python
df.isnull().sum()
```

Remove Duplicate Rows

```python
df = df.drop_duplicates()
```

Convert Price to Numeric

```python
df["price"] = df["price"].astype(str)
df["price"] = df["price"].str.replace(",", "")
df["price"] = pd.to_numeric(df["price"], errors="coerce")
```

Convert Ratings to Numeric

```python
df["ratings"] = pd.to_numeric(df["ratings"], errors="coerce")
```

Remove imgURL column

```python
df = df.drop(columns=["imgURL"], errors="ignore")
```

```python
import pandas as pd

df_clean = pd.read_csv('/content/samsung_mobiles_clean.csv')

data_dict = []
for col in df_clean.columns:
    col_info = {
        'column_name': col,
        'data_type': str(df_clean[col].dtype),
        'non_null_count': df_clean[col].count()
    }
    if df_clean[col].dtype == 'object': # Categorical or string column
        unique_values = df_clean[col].nunique()
        top_5_values = df_clean[col].value_counts().head(5).index.tolist()
        col_info['unique_values_count'] = unique_values
        col_info['top_5_unique_values'] = top_5_values
    else: # Numeric column
        col_info['mean'] = df_clean[col].mean()
        col_info['median'] = df_clean[col].median()
        col_info['std'] = df_clean[col].std()
        col_info['min'] = df_clean[col].min()
        col_info['max'] = df_clean[col].max()
    data_dict.append(col_info)

data_dictionary_df = pd.DataFrame(data_dict)
print("Data Dictionary for samsung_mobiles_clean.csv:")
display(data_dictionary_df)
```

Handle Missing Values

```python
df["ratings"].fillna(df["ratings"].median(), inplace=True)

df["ram"].fillna(df["ram"].median(), inplace=True)

df["storage"].fillna(df["storage"].median(), inplace=True)

df["battery"].fillna(df["battery"].median(), inplace=True)
```

Remove Outliers using IQR

```python
Q1 = df["price"].quantile(0.25)
Q3 = df["price"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df = df[(df["price"] >= lower) & (df["price"] <= upper)]
```

Final Dataset Check

Export Clean Dataset

```python
df.to_csv("samsung_mobiles_clean.csv", index=False)
```
